Los gases de efecto invernadero tienen efectos de gran alcance sobre el medio ambiente y la salud. Provocan el cambio climático al atrapar el calor y también contribuyen a las enfermedades respiratorias causadas por el smog y la contaminación atmosférica. El clima extremo, las alteraciones del suministro de alimentos y el aumento de los incendios forestales son otros efectos del cambio climático provocado por los gases de efecto invernadero. Los patrones meteorológicos típicos a los que estamos acostumbrados cambiarán; algunas especies desaparecerán; otras emigrarán o crecerán.

Al atrapar el calor del sol, los gases de efecto invernadero han mantenido el clima de la Tierra habitable para los seres humanos y millones de otras especies. Pero esos gases están ahora desequilibrados, y amenazan con cambiar drásticamente qué seres vivos pueden sobrevivir en este planeta, y dónde.

Prácticamente todos los sectores de la economía mundial, desde la industria manufacturera a la agricultura, pasando por el transporte y la producción de energía, contribuyen con gases de efecto invernadero a la atmósfera, por lo que todos ellos deben alejarse de los combustibles fósiles si queremos evitar los peores efectos del cambio climático.

# "Emisiones de gases de efecto invernadero y contaminantes atmosféricos procedentes de 28 países de la Unión Europea entre los años 2008 a 2018"

El conjunto de datos tiene siete dimensiones:

1) Contaminante atmosférico [AIRPOL]: Se presentan los resultados de 14 emisiones atmosféricas, a saber:

dióxido de carbono (CO2)
hidrofluorocarburos en equivalentes de CO2 (HFC_CO2E)
metano (CH4)
perfluorocarbonos en equivalentes de CO2 (PFC_CO2E)
óxido nitroso (N2O)
óxidos de azufre (SOX) en equivalentes de dióxido de azufre (SO2) (SOX_SO2E)
amoníaco (NH3)
óxidos de nitrógeno (NOX) en equivalentes de dióxido de nitrógeno (NO2) (NOX_SO2E)
monóxido de carbono (CO)
compuestos orgánicos volátiles no metanos (COVNM)
material particulado < 10 μm (PM10)
material particulado < 2,5 μm (PM2_5)
trifluoruro de nitrógeno y hexafluoruro de azufre en equivalentes de CO2 (NF3_SF6_CO2E)
[notificados conjuntamente] De estas emisiones atmosféricas, el dióxido de carbono, el metano, el óxido nitroso, los hidrofluorocarbonos, los perfluorocarbonos, el trifluoruro de nitrógeno y el hexafluoruro de azufre son gases de efecto invernadero.
Además, el conjunto de datos incluye varios contaminantes atmosféricos expresados en equivalentes de otro contaminante atmosférico:

CH4 en equivalentes de CO2 (CH4_CO2E)(*)

N2O en equivalentes de CO2 (N2O_CO2E)

NH3 en SO2 equivalentes (NH3_SO2E)(*)

NOX en SO2 equivalentes (NOX_SO2E)

NOX en equivalentes de COVNM (NOX_NMVOCE)

CO en equivalentes de COVNM (CO_NMVOCE)

CH4 en equivalentes de COVNM (CH4_NMVOCE) (*) Los valores potenciales de calentamiento global utilizados para expresar CH4 y N2O en equivalentesde CO 2 son 25 y 298, respectivamente. Véase el cuadro 2.14 del cuarto informe de evaluación del IPCC (AR4 Climate Change 2007: The Physical Science Basis - Chapter 2, Changes in atmospheric constituents and in radiative forcing). Los contaminantes atmosféricos expresados en equivalentes de otros contaminantes atmosféricos permiten calcular las siguientes presiones ambientales:

Presión ambiental: Potencial de calentamiento global

Código: GEI

Cálculo: CO2 + N2O en CO2 eq. + CH4 en CO2 eq. + HFC en CO2 eq. + PFC en CO2 eq.+ NF3 y SF6 en CO2 eq.

Presión ambiental: Acidificación

Código: ACG

Cálculo: SOX en SO2 eq. + NOX en SO2 eq. + NH3 en SO2 eq.

Presión ambiental: Precursores del ozono troposférico

Código: O3PR

Cálculo: NMVOC + NOX en NMVOC eq + CO en NMVOC eq. + CH4 en NMVOC eq.

Obsérvese que en ciertos casos los gases fluorados (HFC, PFC, NF3 y SF6) han sido colmados por Eurostat.

2) Clasificación estadística de productos por actividad en la Comunidad Económica Europea, versión 2008 [CPA08]: Los productos para uso final, para los que se estiman y presentan las emisiones atmosféricas «incorporadas», se clasifican de acuerdo con la clasificación CPA 2008. El nivel de desagregación del producto es principalmente el nivel de 2 dígitos; en total, 64 grupos de productos. Además, se reportan las emisiones directas de los hogares privados.

3) Entidad geopolítica (informes) [GEO]: Unión Europea

4) Lugar de origen [ORIGEN]: El lugar de donde se supone que procede la emisión atmosférica "incorporada" estimada. Tres valores son posibles:

DOM = «economía doméstica», es decir, se supone que las emisiones se originan en las actividades económicas nacionales (producción y consumo). ROW = «Resto del mundo (estimado: suponiendo tecnologías nacionales)», es decir, se supone que las emisiones proceden del resto de las economías mundiales; por la que se aplica el llamado «supuesto de tecnología nacional». WORLD = 'Total', es decir, la suma de los dos anteriores.

5) Período de tiempo [TIME]: los datos son anuales. 6) Columnas [INDUSE]: Esta dimensión da cabida a las diversas categorías de cuentas nacionales de «uso final», una clasificación de columnas específica aplicada en tablas simétricas input-output y tablas de uso. En un primer nivel de desagregación, el "uso final" total incluye:

Gastos de consumo final (por hogares, gobierno, etc.);

Formación bruta de capital (principalmente bienes de inversión utilizados para la formación de capital, como, por ejemplo, edificios, equipos de maquinaria, etc.); Exportaciones (es decir, bienes y servicios vendidos a otras economías del resto del mundo). Obsérvese que el concepto «huella» generalmente capta las emisiones causadas por la demanda final interna en la economía de referencia; en otras palabras, causada por el gasto en consumo final y la formación bruta de capital. Las emisiones incorporadas en las exportaciones de la UE cuentan para la huella de otros países y se excluyen de la huella de la UE.

7) Unidad [UNIT]: toneladas, miles de toneladas y kilogramos per cápita.

**Selección Iterativa:**

*Estos métodos comienzan con un conjunto completo de características y, en cada iteración, eliminan o agregan características en función del rendimiento del modelo. Ejemplos de esto incluyen el algoritmo "Forward Selection" (Selección hacia adelante) y "Backward Elimination" (Eliminación hacia atrás).*

**Forward Selection:**

```python
import statsmodels.api as sm
def forward_selection(data, target, significance_level=0.01):
    initial_features = data.columns.tolist()
    best_features = []
    while (len(initial_features)>0):
        remaining_features = list(set(initial_features)-set(best_features))
        new_pval = pd.Series(index=remaining_features)
        for new_column in remaining_features:
            model = sm.OLS(target, sm.add_constant(data[best_features+[new_column]])).fit()
            new_pval[new_column] = model.pvalues[new_column]
        min_p_value = new_pval.min()
        if(min_p_value<significance_level):
            best_features.append(new_pval.idxmin())
        else:
            break
    return best_features
```

*Forward selection es un tipo de regresión paso a paso que comienza con un modelo vacío y agrega variables una por una. En cada paso hacia adelante, se agrega la variable que brinda la mejor mejora única a su modelo. Es uno de los dos métodos comúnmente utilizados de regresión paso a paso; el otro es la eliminación hacia atrás y es casi lo opuesto.*

**Backward Selection:**

```python
def backward_elimination(data, target,significance_level = 0.05):
    features = data.columns.tolist()
    while(len(features)>0):
        features_with_constant = sm.add_constant(data[features])
        p_values = sm.OLS(target, features_with_constant).fit().pvalues[1:]
        max_p_value = p_values.max()
        if(max_p_value >= significance_level):
            excluded_feature = p_values.idxmax()
            features.remove(excluded_feature)
        else:
            break
    return features
```

*La eliminación hacia atrás es un método de regresión paso a paso que comienza con un modelo que contiene todas las variables y elimina variables una por una. En cada paso hacia atrás, se elimina la variable que brinda la peor mejora única a su modelo. Es uno de los dos métodos comúnmente utilizados de regresión paso a paso; el otro es la selección hacia adelante.*

**Eliminación bidireccional (stepwise):**

```python
def stepwise_selection(data, target,SL_in=0.05,SL_out = 0.05):
    initial_features = data.columns.tolist()
    best_features = []
    while (len(initial_features)>0):
        remaining_features = list(set(initial_features)-set(best_features))
        new_pval = pd.Series(index=remaining_features)
        for new_column in remaining_features:
            model = sm.OLS(target, sm.add_constant(data[best_features+[new_column]])).fit()
            new_pval[new_column] = model.pvalues[new_column]
        min_p_value = new_pval.min()
        if(min_p_value<SL_in):
            best_features.append(new_pval.idxmin())
            while(len(best_features)>0):
                best_features_with_constant = sm.add_constant(data[best_features])
                p_values = sm.OLS(target, best_features_with_constant).fit().pvalues[1:]
                max_p_value = p_values.max()
                if(max_p_value >= SL_out):
                    excluded_feature = p_values.idxmax()
                    best_features.remove(excluded_feature)
                else:
                    break
        else:
            break
    return best_features
```

*La eliminación bidireccional es un método de regresión paso a paso que combina la selección hacia adelante y la eliminación hacia atrás. Comienza con un modelo vacío y agrega variables una por una. En cada paso hacia adelante, se agrega la variable que brinda la mejor mejora única a su modelo. En cada paso hacia atrás, se elimina la variable que brinda la peor mejora única a su modelo. Es uno de los dos métodos comúnmente utilizados de regresión paso a paso; el otro es la selección hacia adelante y la eliminación hacia atrás.*
